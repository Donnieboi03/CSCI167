```python
def maxpool(x_in):
  x_out = np.zeros(( int(np.floor(x_in.shape[0]/2)), int(np.floor(x_in.shape[1]/2)) ))
  # TODO -- write the maxpool routine
  # Replace this line
  for i in range(x_out.shape[0]):
    for j in range(x_out.shape[1]):
      x_out[i,j] = np.max(x_in[i*2:(i*2)+2, j*2:(j*2)+2])

  return x_out
```

```python
def duplicate(x_in):
  x_out = np.zeros(( x_in.shape[0]*2, x_in.shape[1]*2 ))
  # TODO -- write the duplication routine
  # Replace this line
  for i in range(x_in.shape[0]):
    for j in range(x_in.shape[1]):
      val = x_in[i,j]
      x_out[i*2:(i*2)+2, j*2:(j*2)+2] = val

  return x_out
```

```python
def max_unpool(x_in, x_high_res):
  x_out = np.zeros(( x_in.shape[0]*2, x_in.shape[1]*2 ))
  # TODO -- write the subsampling routine
  # Replace this line
  for i in range(x_in.shape[0]):
    for j in range(x_in.shape[1]):
      val = x_in[i,j]
      # Get the 2x2 block from the high-resolution image
      block = x_high_res[i*2:(i*2)+2, j*2:(j*2)+2]
      # Find the coordinates where the max value (val) is located in this block
      # Using np.where to get all occurrences, and taking the first one if duplicates exist
      coords_in_block = np.argwhere(block == val)
      if coords_in_block.size > 0:
        r, c = coords_in_block[0] # Take the first occurrence
        x_out[i*2 + r, j*2 + c] = val

  return x_out
```

Finally, we'll try upsampling using bilinear interpolation.  We'll treat the positions off the image as zeros by padding the original image and round fractional values upwards using np.ceil()

```python
def bilinear(x_in):
  x_out = np.zeros(( x_in.shape[0]*2, x_in.shape[1]*2 ))
  x_in_pad = np.zeros((x_in.shape[0]+1, x_in.shape[1]+1))
  x_in_pad[0:x_in.shape[0],0:x_in.shape[1]] = x_in
  # TODO -- write the duplication routine
  # Replace this line
  for i in range(x_out.shape[0]):
    for j in range(x_out.shape[1]):
      # Calculate the corresponding coordinates in the input image
      # Divide by 2 because the output is twice the size of the input
      # And subtract 0.5 to center the sampling points
      x_coord_in = i / 2.0 - 0.5
      y_coord_in = j / 2.0 - 0.5

      # Get the integer and fractional parts of the coordinates
      x_int = int(np.floor(x_coord_in))
      y_int = int(np.floor(y_coord_in))
      x_frac = x_coord_in - x_int
      y_frac = y_coord_in - y_int

      # Ensure coordinates are within bounds for x_in_pad
      x1 = max(0, min(x_int + 1, x_in_pad.shape[0] - 1))
      x2 = max(0, min(x_int + 2, x_in_pad.shape[0] - 1))
      y1 = max(0, min(y_int + 1, x_in_pad.shape[1] - 1))
      y2 = max(0, min(y_int + 2, x_in_pad.shape[1] - 1))

      # Get the four surrounding pixel values from the padded input
      P11 = x_in_pad[x1, y1]
      P12 = x_in_pad[x1, y2]
      P21 = x_in_pad[x2, y1]
      P22 = x_in_pad[x2, y2]

      # Perform bilinear interpolation
      x_out[i,j] = P11 * (1 - x_frac) * (1 - y_frac) + \
                   P21 * x_frac * (1 - y_frac) + \
                   P12 * (1 - x_frac) * y_frac + \
                   P22 * x_frac * y_frac

  return x_out
```
